```python

listings_df["price_float"] = listings_df["price"].str.replace("[$,]", "", regex=True).astype(float)
listings_df["price_float"]
```

```python
plt.figure(figsize=(20, 5))
plt.hist(listings_df["price_float"], bins=50)
plt.title("Price distribution")
plt.plot()
```

* This attribute has a extreme outlier making difficult a proper visualization?

Rta:
In the graph, we observe that the distribution of housing prices has a particular behavior in the first quartile. Starting from the left end of the x-axis, where the price is at 0, there are no homes listed, indicating that there is a significant absence of properties at low prices.

However, as we move further along the x-axis and reach a home price of around 25,000, there is a dramatic change in the distribution. At this point, the number of homes listed increases abruptly, exceeding 100 in that price range.

This pattern suggests that viewing homes in the 0 to 25,000 price range can be complicated by a lack of properties in the lower price ranges and a significant concentration in the 25,000 to 125,000 range. It is important to keep this distribution in mind when analyzing home price data.

```python
q1 = listings_df["price_float"].quantile(0.25)
q3 = listings_df["price_float"].quantile(0.75)
iqr = q3 - q1
```

```python
plt.figure(figsize=(20, 5))
# Calcula los valores Q1, Q3 e IQR
q1 = listings_df["price_float"].quantile(0.25)
q3 = listings_df["price_float"].quantile(0.75)
iqr = q3 - q1

# Define los límites para el eje x
x_ticks = range(0, int(q3 + 1.5 * iqr) + 1, 50)

# Crea el histograma con los límites personalizados en el eje x
plt.figure(figsize=(20, 5))
plt.hist(listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)]["price_float"], bins=25)
plt.title("Price distribution")
plt.xlabel("Price Range")
plt.ylabel("Number of Listings")
plt.xticks(x_ticks) 
plt.show()
```

```python
listings_df["has_availability"] = listings_df["has_availability"].replace({"t": "has_availability", "f": "has_not_availability"})
listings_df["instant_bookable"] = listings_df["instant_bookable"].replace({"t": "instant_bookable", "f": "not_instant_bookable"})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(20, 5))

# Gráfico 1: "has_availability"
# crea figura,con dos graficos, 1 fila dos columnas, uno al lado del otro
has_availability_counts = listings_df["has_availability"].value_counts()
has_availability_counts.plot(kind="bar", ax=ax1, color="orange")
ax1.set_title("Has availability frecuency")
ax1.set_xlabel("Availability")
ax1.set_ylabel("Number of Listings")

ax1.set_xticklabels(has_availability_counts.index, rotation=0, fontsize=16)

# Gráfico 2: "instant_bookable"
# se ordena en orden decendente y cuenta cada categoria de instant_bookable
instant_bookable_counts = listings_df["instant_bookable"].value_counts().sort_index(ascending=False)
instant_bookable_counts.plot(kind="bar", ax=ax2, color="green")
ax2.set_title("Instant bookable frecuency")
ax2.set_xlabel("Instantly Bookable")

# Ajusta las etiquetas del eje x para el segundo subgráfico
ax2.set_xticklabels(instant_bookable_counts.index, rotation=0, fontsize=16)

# Muestra los gráficos
plt.show()
```

```python
# Define los intervalos deseados en el eje x (de 0 a un valor máximo, en incrementos de 50)
x_ticks = range(0, listings_df["number_of_reviews"].max() + 51, 50)

plt.figure(figsize=(20, 5))

# diagrama de caja para la columnam
# showmeans=True mostrara el marcador para la media del diagrama de caja
# va formar la caja de forma horizontal vert=False
plt.boxplot(listings_df["number_of_reviews"], showmeans=True, vert=False)
plt.title("Number of reviews distribution")

# Establece los límites personalizados en el eje x
plt.xticks(x_ticks)

plt.show()
```

```python
listings_sample_df =  listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)].sample(frac=0.1)
listings_sample_df.shape
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x=listings_sample_df["price_float"],
    y=listings_sample_df["review_scores_cleanliness"]
)
plt.title("Relationship between price and review score cleanliness")
plt.grid()
plt.show()
```

```python
neighbourhood_frec_cumsum = listings_df["neighbourhood_cleansed"].value_counts(normalize=True).cumsum()
```

```python
# Calcula el porcentaje acumulado
neighbourhood_frec_cumsum_percent = neighbourhood_frec_cumsum * 100

plt.figure(figsize=(20, 8))
neighbourhood_frec_cumsum_percent.plot(kind="bar", color="steelblue")
plt.title("Pareto analysis for neighbourhood frecuency")
plt.ylabel("Percentage (%)")  # Agrega etiqueta al eje y
plt.grid(axis="y")
plt.show()
```

```python
most_representative_neighbourhoods = neighbourhood_frec_cumsum.loc[neighbourhood_frec_cumsum < 0.8].index.tolist()
most_representative_neighbourhoods 
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 8))
sns.kdeplot(
    data=listings_df.loc[listings_df["price_float"] <= (q3 + 1.5 * iqr)].loc[listings_df["neighbourhood_cleansed"].isin(most_representative_neighbourhoods)],
    x="price_float",
    hue="neighbourhood_cleansed",
    bw_adjust=.3,
    ax=ax
)
```
